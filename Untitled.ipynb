```python
from huggingface_hub import hf_hub_download
from llama_cpp import Llama
from transformers import AutoTokenizer

# 1. Update model name and file for TinyLlama
model_name = "TheBloke/TinyLlama-1.1B-Chat-v1.0-GGUF"
model_file = "tinyllama-1.1b-chat-v1.0.Q4_K_M.gguf" # A 4-bit quantized version

# Download the model file
print(f"Downloading model: {model_file} from {model_name}...")
model_path = hf_hub_download(repo_id=model_name, filename=model_file)

# Load the model using Llama.cpp
# TinyLlama was trained with a 2048 context length
llm = Llama(model_path=model_path, n_ctx=2048, verbose=False)

# 2. Update the tokenizer to match the TinyLlama model
# GGUF models don't have the tokenizer bundled. We can load it from the original model repo.
try:
    tokenizer = AutoTokenizer.from_pretrained("TinyLlama/TinyLlama-1.1B-Chat-v1.0", use_fast=True)
except Exception as e:
    print(f"Could not load the specific tokenizer: {e}")
    # Fallback to a generic tokenizer if the specific one fails
    tokenizer = AutoTokenizer.from_pretrained("gpt2", use_fast=True)

# --- Example Usage ---

```

```python
from langchain_community.llms import LlamaCpp
from langchain.chains import RetrievalQA

# Load the model using Llama.cpp
# model_path is already defined in the previous cell
llm = LlamaCpp(
    model_path=model_path,
    temperature=0.7,
    max_tokens=512,
    top_p=0.95,
    n_ctx=2048 # Set the context window size
)
```

```python
from langchain.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS

# Load the PDF
loader = PyPDFLoader("1.pdf")
docs = loader.load()

# Split into chunks
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=50)
chunks = text_splitter.split_documents(docs)

# Create embeddings and vector store
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
db = FAISS.from_documents(chunks, embeddings)

```

```python
from langchain.prompts import PromptTemplate

template = """
You are a helpful assistant with the following persona: {persona}

Your task is: {task}

Use the following context to answer:
{context}

Answer:
"""

prompt = PromptTemplate(
    input_variables=["persona", "task", "context"],
    template=template,
)
```

```python
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.chains import RetrievalQA

# Custom prompt
template = """
You are a helpful assistant with the following persona: {persona}

Your task is: {task}

Use the following context to answer:
{context}

Answer:
"""

prompt = PromptTemplate(
    input_variables=["persona", "task", "context"],
    template=template,
)
from langchain.chains import LLMChain

# This will handle the final generation step
qa_llm_chain = LLMChain(llm=llm, prompt=prompt)
```

```python
# 1. Retrieve docs using "task" as the query
retrieved_docs = db.similarity_search("Plan a trip of 4 days for a group of 10 college friends.")
context = "\n\n".join([doc.page_content for doc in retrieved_docs])

# 2. Run LLM chain with all inputs
response = qa_llm_chain.run({
    "persona": "Travel Planner",
    "task": "Plan a trip of 4 days for a group of 10 college friends.",
    "context": context,
})

print(response)
```
